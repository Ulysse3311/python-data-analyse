```python
import selenium
from selenium import webdriver
import numpy as np
import pandas as pd
path   = "chromedriver.exe"
```

Accès au lieu

```python
driver = webdriver.Chrome(executable_path=path)
url= "https://aqicn.org/data-platform/register/fr/"
driver.get(url)
driver.implicitly_wait(10)
lieu="Palais De Justice, Rouen, France"
driver.find_element_by_class_name("prompt").send_keys(lieu)
driver.find_element_by_xpath("/html/body/div[7]/center[3]/div/div[2]/div[1]/div[2]/div[2]/a[1]/div/div").click()
```

Parcour du tableau de données 

```python
tableau=driver.find_element_by_xpath("/html/body/div[7]/center[3]/div/div[2]/div[3]/div[2]/center/div[2]/table/tbody")
m=12
j=0
mois=""
expdate=[]
expdata=[]
for x in tableau.find_elements_by_xpath(".//tr")[2:13]:
    if x.find_element_by_xpath(".//td[1]").text != mois:
        m-=1
    listedata=x.find_element_by_xpath(".//td[4]").text.split("\n")
    j=1
    for data in listedata:
        expdate.append(f"{j}-{m}-2020")
        if data=="-":
            data=np.nan
        expdata.append(data)
        j+=1
    
#print(expdata)
#print(expdate)
```

Création du dataframe et visualisation

```python
df=pd.DataFrame({"date":expdate,"PM2.5":expdata})
df["date"]=pd.to_datetime(df["date"],dayfirst=True)
df["PM2.5"]=df["PM2.5"].astype(float)
df=df.set_index("date")
df.plot();
```
